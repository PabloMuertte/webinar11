```python
from sympy import *
```

- Функция (f_x)

```python
x = Symbol('x')
f_x = sympify('- 18*x^3 + 5*x^2 + 10*x - 30') #-12*x^4*sin(cos(x)) - 18*x^3 + 5*x^2 + 10*x - 30       8*x^3 -36*x^2 + 54*x -27
f_x
```

- Корни функции

```python
roots = solve(f_x)
print(f'Корни:', roots)
```

- График функции

```python
chart_1 = plotting.plot(f_x, (x, -1, 1), line_color='red')
```

- Интервалы, на которых функция возрастает

```python
print(solve(f_x > 0))
```

- Интервалы, на которых функция убывает

```python
print(solve(f_x < 0))
```

<br/><br/><br/>
- Производная функции (derivative_f_x)

```python
derivative_f_x = f_x.diff(x)
derivative_f_x
```

- График производной функции

```python
chart_2 = plotting.plot(derivative_f_x, (x, -0.5, 0.5), line_color='green')
```

- Вершина в точке 

```python
print(solve(derivative_f_x)[0].evalf())
```

```python
print(solve(derivative_f_x))
```

- Интервалы, на которых производная функции возрастает 

```python
print(solve(derivative_f_x > 0))
```

- Интервалы, на которых производная функции убывает

```python
print(solve(derivative_f_x < 0))
```
